```python
    def _get_locations( m, n):
        """
        Return the indices of an m by n array.
        """
        # the element in these indices are non-zero
        #return a group of indices for each suitable elements in a group or matrix 
        #print("m n {} {}".format(m,n))
        #print("np.ones(shape=(m, n){}".format(np.ones(shape=(m, n))))
        #print("_get_locations( m, n){}".format(np.argwhere(np.ones(shape=(m, n))).astype(np.int64)))
        return np.argwhere(np.ones(shape=(m, n))).astype(np.int64)
```

```python
m =3
n =3
dim =4
lr = 1
sigma =1
max_iter = 3000
random_max_iter  = 100
random_state = 1234
trained = False
shape = (m, n)
initial_lr = lr
rng = np.random.default_rng(random_state)
weights = rng.normal(size=(m * n, dim)) #******** m*n*dim matrix get randomized samples from a normal distrubition
_locations = _get_locations(m, n)
#print(weights)
 # Set after fitting
_inertia = None
_n_iter_ = None
_trained = False

weights1= rng.normal(size=(m * n, dim))

weights2= rng.normal(size=(m * n, dim))
```

```python
    def _find_bmu(x, newWeights):
        """
        Find the index of the best matching unit for the input vector x.
        """  
        #print("x:") 
        #print(x)
        # Stack x to have one row per weight *********** get the all the element for one row
        x_stack = np.stack([x]*(m*n), axis=0)
        # Calculate distance between x and each weight  ， it use the norm to represent the distance of the concept of vector x_stack - newWeights
        #print(weights.shape)
        #print("x_stack - newWeights:{}".format(x_stack - newWeights ))   #, axis =1  process by row
        distance = np.linalg.norm(x_stack - newWeights, axis=1)
        #print("distance:{}".format(distance ))   
       # print("min distance:")   
        #print(np.argmin(distance))

        # Find index of best matching unit
        return np.argmin(distance)
```

```python
    def step(x):
        """
        Do one step of training on the given input vector.
        """
        #print(x)
        # Stack x to have one row per weight 
        x_stack = np.stack([x]*(m*n), axis=0)
        # x_stack , with mxn row , each row has the same array: x
        global weights

        # Get index of best matching unit
        bmu_index = _find_bmu(x,weights)
        #print("bmu_index{}".format(bmu_index));
        # Find location of best matching unit, _locations is all the indices for a given matrix for array
        # bmu_location is the bmu_indexth element in _locations, such as if bmu_index = 4 in [[0,0],[0,1],[1,0],[1,1],[2,0],[2,1]] it return [2,0]
        bmu_location = _locations[bmu_index,:]
        #print("bmu_location{}".format(bmu_location));
        # Find square distance from each weight to the BMU
        #print("[bmu_location]*(m*n){}".format([bmu_location]*(m*n)));
        stacked_bmu = np.stack([bmu_location]*(m*n), axis=0)
        #print("stacked_bmu: {}".format(stacked_bmu))
        #the distance among unit is calcuated by the distance among unit's indices
        #bmu_distance is an array with distance to each unit
        bmu_distance = np.sum(np.power(_locations.astype(np.float64) - stacked_bmu.astype(np.float64), 2), axis=1)
       # print("bmu_distance:{}".format(bmu_distance))
        # Compute update neighborhood
        neighborhood = np.exp((bmu_distance / (sigma ** 2)) * -1)
       # print("neighborhood:{}".format(neighborhood))
        #local_step is an array with stepchanges to each unit
        local_step = lr * neighborhood
       # print("local_step:{}".format(local_step))
        # Stack local step to be proper shape for update
        local_multiplier = np.stack([local_step]*(dim), axis=1)
        #print("local_multiplier:{}".format(local_multiplier))
        #print("local_multiplier:{}".format(local_multiplier))
        # Multiply by difference between input and weights
        delta = local_multiplier * (x_stack - weights)
        #print("delta:{}".format(delta))
        # Update weights
        weights += delta
        #print("weights:{}".format(weights))
```

```python
def _compute_point_intertia( x):
        """
        Compute the inertia of a single point. Inertia defined as squared distance
        from point to closest cluster center (BMU)
        """
        
        # Find BMU
        bmu_index = _find_bmu(x,weights)
        bmu = weights[bmu_index]
        # Compute sum of squared distance (just euclidean distance) from x to bmu
        return np.sum(np.square(x - bmu))
```

```python
def fit( X, weightIndex = 0,epochs=1, shuffle=True):
        """
        Take data (a tensor of type float64) as input and fit the SOM to that
        data for the specified number of epochs.
        Parameters
        ----------
        X : ndarray
            Training data. Must have shape (n, self.dim) where n is the number
            of training samples.
        epochs : int, default=1
            The number of times to loop through the training data when fitting.
        shuffle : bool, default True
            Whether or not to randomize the order of train data when fitting.
            Can be seeded with np.random.seed() prior to calling fit.
        Returns
        -------
        None
            Fits the SOM to the given data but does not return anything.
        """
        # Count total number of iterations
        global_iter_counter = 0
    # the number of samples   
        n_samples = X.shape[0] 
        total_iterations = np.minimum(epochs * n_samples, max_iter)
        for epoch in range(epochs):
            # Break if past max number of iterations
            if global_iter_counter > max_iter:
                break

            if shuffle:
                rng = np.random.default_rng(random_state)
                indices = rng.permutation(n_samples)
                # permute the index of samples
                indices = np.array(indices)
            else:
                indices = np.arange(n_samples)                       
            

         # Train
            for idx in indices:

             # Break if past max number of iterations
                if global_iter_counter > max_iter:
                    break
                #result =  "{} --- {} ".format(i, idx)
                #print(result )
                input = X[idx]
                # Do one step of training
                step(input)
                # Update learning rate
                global_iter_counter += 1
                lr = (1 - (global_iter_counter / total_iterations)) * initial_lr
    
        # Compute inertia
          
        inertia = np.sum(np.array([float(_compute_point_intertia(x)) for x in X]))
        _inertia_ = inertia

    # Set n_iter_ attribute
        _n_iter_ = global_iter_counter

    # Set trained flag
        global trained
        trained = True
        global weights1
        global weights2
        if(weightIndex == 0):
            weights1 = copy.deepcopy(weights)
            print("generate weight1:\n {}".format(weights1))
        if(weightIndex == 1):
            weights2 = copy.deepcopy(weights)
            print("generate weight2:\n {}".format(weights2))
      
        return
```

```python
    def predict(X, newWeights):
        """
        Predict cluster for each element in X.
        Parameters
        ----------
        X : ndarray
            An ndarray of shape (n, self.dim) where n is the number of samples.
            The data to predict clusters for.
        Returns
        -------
        labels : ndarray
            An ndarray of shape (n,). The predicted cluster index for each item
            in X.
        """

        print("weights used:\n")
        print(newWeights)
        # Check to make sure SOM has been fit
        if not trained:
            raise NotImplementedError('SOM object has no predict() method until after calling fit().')

        # Make sure X has proper shape
        assert len(X.shape) == 2, f'X should have two dimensions, not {len(X.shape)}'
        assert X.shape[1] == dim, f'This SOM has dimesnion {dim}. Received input with dimension {X.shape[1]}'
     
        labels = np.array([_find_bmu(x,newWeights) for x in X])
    
        print("predicted label:\n {}".format(labels))
        return labels
```

```python
def NormalizeLables(X,Y,category = 0):
    """
    transfrer predcited label to match the range of true label
    a = np.amax(X)+1,b = np.amax(Y)+1 the number of classes in X and Y
    b= m*n,  to make it easier , to make b can be divided by a
    category = 0 noramlize label
    category = 1 normalize sub label
    """
    div = int((np.amax(Y)+1)/(np.amax(X)+1))
    #print("div : {}".format(div) )
    if(category == 0):
        global nLabel 
        nLabel = np.arange(Y.size)
    if(category == 1):
        global nsubLabel 
        nsubLabel = np.arange(Y.size)
    index = 0;
    for idx, y in enumerate(Y): 
       # print("idx {}".format(idx))
        for i in range(1,div+1):
            if(y < i*div):
                index+=1
                #print(i-1)
                if(category == 0):
                    nLabel[idx] = i-1
                if(category == 1):
                    nsubLabel[idx] = i-1
                break               
    if(category == 0):
        print("normalized predicted nLabel:\n {} ".format(nLabel))
    if(category == 1):
        print("normalized predicted nsubLabel: \n{} ".format(nsubLabel))
    return 
         
```

```python
def purity_score(y_true, y_pred):
    # compute contingency matrix (also called confusion matrix)
    #print(y_true.shape)
    #print(y_pred.shape)
    contingency_matrix = metrics.cluster.contingency_matrix(y_true, y_pred)
    # return purity
    print (np.sum(np.amax(contingency_matrix, axis=0)) / np.sum(contingency_matrix))
    return np.sum(np.amax(contingency_matrix, axis=0)) / np.sum(contingency_matrix) 
```

```python
def calculate_metrics(gt, pred):
    """
    :param gt: real data labels，such as:y=[1,0,1,0,1]
    :param pred: the predicited labels，when calculating confusion matrix, it must be integer, so float needs to change to int first
    :return: the valuation matrix value
    """
    """
        confusion_matrix(y_true,y_pred,labels,sample_weight,normalize)
        y_true:real label；
        y_pred: predicted labels (converted from predicted probability) 
        labels:用于标签重新排序或选择标签子集；
        sample_weight:样本权重；
        normalize:在真实（行）、预测（列）条件或所有总体上标准化混淆矩阵；
    """
    print("starting!!!-----------------------------------------------")
    # the default theme of seaborn
    sns.set()
    fig, (ax1, ax2) = plt.subplots(figsize=(10, 8), nrows=2)
    confusion = confusion_matrix(gt, pred)
    # 打印具体的混淆矩阵的每个部分的值
    print(confusion)
    # 从左到右依次表示TN、FP、FN、TP
    # TN, FP, FN, TP = confusion.ravel()
    print(confusion.ravel())
    # draw confusion matrix 
    sns.heatmap(confusion,annot=True)
    ax2.set_title('sns_heatmap_confusion_matrix')
    ax2.set_xlabel('y_pred')
    ax2.set_ylabel('y_true')
    fig.savefig('sns_heatmap_confusion_matrix.jpg', bbox_inches='tight')
    TP = confusion[1, 1]
    TN = confusion[0, 0]
    FP = confusion[0, 1]
    FN = confusion[1, 0]
    # Get each evaluation indicators through confusion matrix 
   # print('AUC:',roc_auc_score(gt, pred))
    print('Accuracy:', (TP + TN) / float(TP + TN + FP + FN))
    print('Sensitivity:', TP / float(TP + FN))
    print('Specificity:', TN / float(TN + FP))
    print('PPV:',TP / float(TP + FP))
    print('Recall:',TP / float(TP + FN))
    print('Precision:',TP / float(TP + FP))
    # F1-score = 2*recall*precision/recall+precision
    P = TP / float(TP + FP)
    R = TP / float(TP + FN)
    print('F1-score:',(2*P*R)/(P+R))
    print('True Positive Rate:',round(TP / float(TP + FN)))
    print('False Positive Rate:',FP / float(FP + TN))
    print('Ending!!!------------------------------------------------------')
 
    # 采用sklearn提供的函数验证,用于对比混淆矩阵方法与这个方法的区别
    #print("the result of sklearn package")
    #auc = roc_auc_score(gt,pred,multi_class='ovr')
    #print("sklearn auc:",auc)
    #accuracy = accuracy_score(gt,pred)
    #print("sklearn accuracy:",accuracy)
    #recal = recall_score(gt,pred)
    #precision = precision_score(gt,pred)
    #print("sklearn recall:{},precision:{}".format(recal,precision))
    #print("sklearn F1-score:{}".format((2*recal*precision)/(recal+precision)))
```

```python
def groupClusterList(m,Y):
    """
    transfer all label in a list
    m: cluster Number
    """
    clusters = []
    for i in range(0,m):
        newlist = []
        for idx, y in enumerate(Y): 
            if(y == i):
                newlist.append(idx)
        clusters.append(newlist) 
   # print(clusters)
    return clusters
```

```python
def getNoisyClusters(list_true,list_pred):
    """
    get the wrong data indices in the training data when predicted with weight1
    """
    noisylist = []
    for i in range(0,len(list_true)):
        newlist = [item for item in list_pred[i] if item not in list_true[i]]
        noisylist.append(newlist)
    print("noisy data:")
    return noisylist
```

```python
def reduce_noisy_data(noisy_list, percent):
    newlist = []
    for x in noisy_list:
        for e in x:
            newlist.append(e)
   # print(newlist)
   # print(len(newlist))
   # print(int(percent * len(newlist)))
    newlist = random.sample(newlist, int(percent * len(newlist)))
   # print(newlist)
    return newlist
```

```python
def randomprocessdataset(nsamples,predictedLabel,weight,category):
    purity = 0
    for i  in range(0,100):
        subset = np.random.choice(number_of_rows,nsamples,replace = False)
        subset = np.sort(subset)    
        subset1 = iris_data[subset, :]
        subset1_label = iris_label[subset]
        predictions_W_random = predict(subset1,weight)
        NormalizeLables(subset1_label,predictions_W_random,category)
        purity += purity_score(subset1_label,predictedLabel)
        i = i+1
    return purity/100
```
