```python
#THIS CODE ONLY DO THE WHOLE DATA TRAINING, NO CROSS_VALIDATION AND FEATURE SELECTION
import pandas as pd
import numpy as np
import newSom
import optimizeW_CM_MultipleTraining
import dataset_read
import researchpy as rp
import scipy.stats as stats
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning) 
```

```python
dataread = dataset_read.DATAREAD()
```

```python
csv_training = pd.read_csv("Crowdsourced Mapping/training.csv")
csv_test = pd.read_csv("Crowdsourced Mapping/testing.csv")
dataread.stringToIntCrowdsourcedMappingDataSet(csv_training,"class")
dataread.stringToIntCrowdsourcedMappingDataSet(csv_test,"class")
```

```python
som = newSom.SOM(m= 6, n= 2, dim=28)  # dim = i-1, as the first feature is the class label
optimize_W1 = optimizeW_CM_MultipleTraining.OptimizeW(som,csv_training,csv_test,6)
optimize_W1.runOptimize(1,0)
```

```python
# multiple text by adding neuron number - Purity  split number = 2

max_num = 30
unit_list = []    
all_train_score_W0 =[]
all_train_score_W_combine =[]
test_score_W0 = []
test_score_W_combine= []
scoretype = 0
x = 1
while x <= max_num:
    unit_list.append(6*x)
    print("neuron unit number: {}".format(6*x))
    print("*******************\n")
    som = newSom.SOM(m= 6, n= x, dim=28)  # dim = i-1, as the first feature is the class label
    optimize_W = optimizeW_CM_MultipleTraining.OptimizeW(som,csv_training,csv_test,6)
    optimize_W.runOptimize(2,scoretype)
    all_train_score_W0.append(optimize_W.all_train_score_W0)
    all_train_score_W_combine.append(optimize_W.all_train_score_W_Combined)
    test_score_W0.append(optimize_W.test_score_W0)
    test_score_W_combine.append(optimize_W.test_score_W_combined)
    x = x+1

    
if scoretype == 0:
    plt.title("Purity Score")
elif scoretype == 1:
     plt.title("NMI Score")
elif scoretype == 2:
     plt.title("ARI Score")
        
plt.xlabel('Neuron number')
plt.plot(unit_list,all_train_score_W0,'r',label ='all_train_score_W0')
plt.plot(unit_list,all_train_score_W_combine,'c',label ='all_train_score_W\'')
plt.plot(unit_list,test_score_W0,'y',label ='test_score_W1')
plt.plot(unit_list,test_score_W_combine,'k',label ='test_score_W\'')
plt.legend()
plt.show()


df1 = pd.DataFrame(test_score_W0, columns = ['test_score_W0'])
df2 = pd.DataFrame(test_score_W_combine, columns = ['test_score_W\''])

summary, results = rp.ttest(group1= df1['test_score_W0'], group1_name= "test_score_W0",
                            group2= df2['test_score_W\''], group2_name= "test_score_W\'")
print(summary)
print(results)

stats.ttest_ind(test_score_W0, test_score_W_combine,alternative = 'less')
```
